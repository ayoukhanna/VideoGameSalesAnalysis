```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/videogames_db')
```

```python
#load raw data
raw_data = pd.read_csv('./Resources/Video_Games_Sales_as_at_22_Dec_2016.csv')
raw_data
```

```python
#drop nan values
#reset index
no_na = raw_data.dropna()
cleaned_df = no_na.reset_index()
final_df = cleaned_df.drop(['index'],axis=1)
final_df = final_df.rename(columns =str.lower)
final_df
```

```python
final_df.info()
```

```python
#Export to sql
final_df.to_sql(name='videogame_sales',
                        con=engine,
                        index=False,
                        if_exists='append')
```
